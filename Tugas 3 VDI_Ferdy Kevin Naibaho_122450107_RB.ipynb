### Dataset tentang Pemetaan Daerah Banjir

Tujuan: Tujuan dari visualisasi ini adalah untuk memantau ketinggian air di berbagai titik pengamatan dan memahami tren perubahan ketinggian air yang bisa menyebabkan banjir. Hal ini bertujuan untuk memberikan peringatan dini terhadap kemungkinan terjadinya banjir di area tersebut.

User: User visualisasi ini adalah BPBD (Badan Penanggulangan Bencana Daerah), pemerintah daerah, warga yang tinggal di daerah rawan banjir, serta peneliti yang bekerja dalam bidang mitigasi bencana.

Fungsi: Fungsi dari visualisasi ini adalah untuk menyajikan data secara visual agar memudahkan identifikasi kenaikan ketinggian air yang mungkin menjadi ancaman banjir, serta mendukung pengambilan keputusan terkait tindakan mitigasi yang perlu dilakukan.

Tone: Tone dari visualisasi ini bersifat informatif, waspada, dan edukatif. Visualisasi ini berusaha menyampaikan data dengan jelas untuk memberikan pemahaman terhadap kondisi yang mungkin menjadi ancaman.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
flood_data = pd.read_csv('Pemetaan Daerah Banjir.csv')

# Mengonversi kolom 'Tanggal' dan 'Waktu' menjadi satu kolom datetime
flood_data['Datetime'] = pd.to_datetime(flood_data['Tanggal'] + ' ' + flood_data['Waktu'], format='%d/%m/%Y %H:%M:%S')

# Mengatur kolom datetime sebagai index
flood_data.set_index('Datetime', inplace=True)

# Menghitung frekuensi dari status banjir
flood_data['Status Banjir'] = flood_data['Status Banjir'].replace({0: 'Tidak Banjir', 1: 'Siaga 1', 2: 'Siaga 2', 3: 'Siaga 3'})
status_counts = flood_data['Status Banjir'].value_counts()

# Visualisasi 1: Line Plot untuk Perubahan Ketinggian Air
plt.figure(figsize=(14, 8))
sns.lineplot(data=flood_data, x=flood_data.index, y='Katulampa', label='Katulampa')
sns.lineplot(data=flood_data, x=flood_data.index, y='Pos Depok', label='Pos Depok')
sns.lineplot(data=flood_data, x=flood_data.index, y='Manggarai', label='Manggarai')

plt.title('Perubahan Ketinggian Air di Berbagai Titik Pengamatan Seiring Waktu', fontsize=16)
plt.xlabel('Waktu', fontsize=14)
plt.ylabel('Ketinggian Air (cm)', fontsize=14)
plt.legend(title='Titik Pengamatan', loc='upper right')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualisasi 2: Bar Plot untuk Frekuensi Status Banjir
plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, hue=status_counts.index, palette='Blues', dodge=False)
plt.legend([], [], frameon=False)

plt.title('Frekuensi Status Banjir', fontsize=16)
plt.xlabel('Status Banjir', fontsize=14)
plt.ylabel('Frekuensi', fontsize=14)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

 Kode ini menghasilkan dua jenis visualisasi: line plot untuk memantau ketinggian air di beberapa titik pengamatan, dan bar plot untuk melihat frekuensi status banjir.
